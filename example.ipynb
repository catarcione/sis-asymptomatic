# Importing libraries and modules

```python
import epidemic
import utils
import networkx as nx
```

# Setting parameters

```python
network = nx.barabasi_albert_graph(3000, 4)

beta = 0.1 # Transmission probability
gamma = 0.08 # Recovery probability
asymp_frac = 0.2 # Fraction of asymptomatic nodes

initial_infected_count = 50
burn_in_steps = 100
num_snapshots = 5
steps_between = 8

prediction_fractions = [0.1, 0.5, 0.9]
```

# Designating asymptomatic nodes and running epidemic simulation

```python
asymptomatic = epidemic.choose_asymptomatic_nodes(network, asymp_frac)

infected = epidemic.burn_in(network, beta, gamma, initial_infected_count, burn_in_steps)

snapshots = epidemic.collect_snapshots(network, beta, gamma, infected, num_snapshots, steps_between)
```

# Computing network centrality measures

```python
observed_betws, contacts, observations = utils.compute_centrality_metrics(network, snapshots, asymptomatic)

print("Centrality metrics computed for all snapshots")
for t in range(len(snapshots)):
    print(f"  Time {t}: {len(observations[t])} total observed infected nodes")
```

# Evaluating prediction methods

```python
naive_scores, obs_betw_scores, contact_scores = utils.evaluate_prediction_methods(
    network,
    snapshots,
    asymptomatic,
    observed_betws,
    contacts,
    observations,
    prediction_fractions
)
```

```python
# Naive baseline results
print("=" * 60)
print("NAIVE BASELINE (predict all unobserved as asymptomatic)")
print("=" * 60)
for time_key, metrics in naive_scores.items():
    print(f"\n{time_key}:")
    print(f"  Precision: {metrics['precision']:.3f}")
    print(f"  Recall:    {metrics['recall']:.3f}")
    print(f"  F1 Score:  {metrics['f1']:.3f}")
```

```python
# Observed betweenness results
print("\n" + "=" * 60)
print("OBSERVED BETWEENNESS CENTRALITY")
print("=" * 60)
for time_key, frac_results in obs_betw_scores.items():
    print(f"\n{time_key}:")
    for frac_key, metrics in frac_results.items():
        print(f"  {frac_key}:")
        print(f"    Precision: {metrics['precision']:.3f}")
        print(f"    Recall:    {metrics['recall']:.3f}")
        print(f"    F1 Score:  {metrics['f1']:.3f}")
```

```python
# Contact score results
print("\n" + "=" * 60)
print("CONTACT SCORE (fraction of infected neighbors)")
print("=" * 60)
for time_key, frac_results in contact_scores.items():
    print(f"\n{time_key}:")
    for frac_key, metrics in frac_results.items():
        print(f"  {frac_key}:")
        print(f"    Precision: {metrics['precision']:.3f}")
        print(f"    Recall:    {metrics['recall']:.3f}")
        print(f"    F1 Score:  {metrics['f1']:.3f}")
```
